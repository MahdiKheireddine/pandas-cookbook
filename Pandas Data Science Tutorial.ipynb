## Making changes to the data

#### Create new column , Drop 

```python
# Calculate the total stats for each Pokémon by summing up the values of their individual stats
# using the provided columns: HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed.
# This creates a new column named 'Total'.
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']


# Drop the previously created 'Total' column to avoid redundancy.
#df = df.drop(columns=['Total'])


# Recalculate the 'Total' column using a more concise approach by summing up the values
# of the specified columns (HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed) across rows.
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

df.head()
```

#### Reorganization of columns

```python
# Get the list of column names
cols = list(df.columns)

# Reorder the columns such that the 'Total' column is moved next to the first four columns
# and the rest of the columns remain in their original order
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

## Filtering Data

```python
# Filter the DataFrame based on Pokémon with 'Type 1' as 'Grass' and 'Type 2' as 'Poison'
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

# Filter the DataFrame further based on Pokémon with 'Type 1' as 'Grass', 'Type 2' as 'Poison', and total stats greater than or equal to 300
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['Total'] >= 300)]

# Filter the DataFrame based on Pokémon with either 'Type 1' as 'Grass' or 'Type 2' as 'Poison'
new_df = df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]

# Reset the index of the filtered DataFrame and modify it in place
new_df.reset_index(drop=True, inplace=True)

# Export the filtered DataFrame to a CSV file named 'filtered.csv'
new_df.to_csv('data\\filtered.csv')

new_df
```

#### Filtering 2

```python
# Filter the DataFrame to include only rows where the 'Name' column contains the substring "Mega"
new_df2 = df.loc[df['Name'].str.contains("Mega")]

# Alternative: Filter the DataFrame to exclude rows where the 'Name' column contains the substring "Mega"
# new_df2 = df.loc[~df['Name'].str.contains("Mega")]

# Reset the index of the filtered DataFrame and modify it in place
new_df2.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
new_df2
```

#### Filtering 3 using regex

```python
import re

# Filter the DataFrame to include rows where 'Type 1' contains "fire" or "Grass" 
new_df3 = df.loc[df['Type 1'].str.contains("fire|Grass", flags=re.I, regex=True)]

# Reset the index of the filtered DataFrame and modify it in place
new_df3.reset_index(drop=True, inplace=True)

# Filter the DataFrame to include rows where 'Name' starts with "pi" followed by any lowercase letters 
new_df4 = df.loc[df['Name'].str.contains("^pi[a-z]*", flags=re.I, regex=True)]

# Reset the index of the filtered DataFrame and modify it in place
new_df4.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
new_df4
```


## Conditional Changes

```python
df_change = df.copy()
# Set the value of the 'Legendary' column to True for rows where the value in the 'Type 1' column is 'Fire'
df_change.loc[df['Type 1'] == 'Fire', 'Legendary'] = True

# Set the value of the 'Mega_Evolution' column to True for rows where the name of the Pokémon contains "Mega"
df_change.loc[df['Name'].str.contains("Mega"), 'Mega_Evolution'] = True

# Set the value of the 'Legendary' column to True for rows where the name contains "Mega" and 'Type 2' is 'Flying'
# Also, set the value of the 'Mega_flyer' column to True for the same rows
df.loc[(df['Name'].str.contains("Mega")) & (df['Type 2'] == 'Flying'), ['Legendary', 'Mega_flyer']] = [True, True]

# Same as above
# Set the value of the 'Legendary' and 'Mega_flyer' columns to True for rows where the name contains "Mega" and 'Type 2' is 'Flying'
df.loc[(df['Name'].str.contains("Mega")) & (df['Type 2'] == 'Flying'), ['Legendary', 'Mega_flyer']] = True

df_change
```

## Aggregate Statistics (Groupby)


```python
df_modif = pd.read_csv('data\modified.csv')

# Group the DataFrame by the values in the 'Type 1' column, calculate the mean for each group
# and sort the result based on the 'Defense' column in descending order
result = df_modif.groupby(['Type 1']).mean(numeric_only=True).sort_values('Defense', ascending=False)

# Group the DataFrame by the values in the 'Type 1' column, calculate the sum for each group
result = df_modif.groupby(['Type 1']).sum()

result
```

## Working with large amounts of data



```python
# Initialize an empty DataFrame with the same columns as df
new_df = pd.DataFrame(columns=df.columns)

# Read the CSV file in chunks of size 5
for chunk in pd.read_csv('data\modified.csv', chunksize=5):
    # Group the chunk by the 'Type 1' column and count occurrences of each group
    results = chunk.groupby(['Type 1']).count()
    
    # Concatenate the results with new_df
    new_df = pd.concat([new_df, results])

new_df
```
